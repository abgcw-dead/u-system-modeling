```python
import math
import matplotlib.pyplot as plt
import numpy as np
# from scipy.integrate import odeint
```

```python
ke = 0.025
temp_env = 22
temp_initial = 83
time_initial = 0
time_max = 1000
time_step = 1
step_n = 100
step_s = 0.001
```

```python
def f(temp):
    return -ke * (temp - temp_env)
```

```python
def euler(n, h, x, y):
    for _ in range(n):
        y += h * f(y)  # f(x, y)
        x += h
    return x, y
```

```python
time = [t for t in range(time_max)]
temp = [temp_initial]
for i in range(len(time)):
    result = euler(step_n, step_s, time[i], temp[i])
    temp.append(result[1])
temp = temp[:len(time)]
```

```python
plt.plot(time, temp)
plt.show()
```
