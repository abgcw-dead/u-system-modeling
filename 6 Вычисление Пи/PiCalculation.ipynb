```python
import matplotlib.pyplot as plt
import random as rnd
import numpy as np
```

```python
n_list = [10 ** i for i in range(8)]
```

```python
def circle(x, y):
    return x ** 2 + y ** 2 <= 1

pi_list = []
for n in n_list:
    x_list = np.random.uniform(0, 1, n)
    y_list = np.random.uniform(0, 1, n)
    # x_list = [rnd.random() for _ in range(n)]
    # y_list = [rnd.random() for _ in range(n)]
    result = circle(x_list, y_list)
    pi_list.append(4 * sum(result) / n)

plt.axhline(y=np.pi, color='red', linestyle=':')
plt.plot(n_list, pi_list)
# plt.ylim(bottom=0)
plt.xscale('log')
plt.xlabel('Количество песчинок')
plt.ylabel('Значение Пи')
plt.grid()
plt.show()
```
