```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
n = 10000

a = 0
b = 5
c = (b - a) / 2

bins_n = 100
```

### Нормальное -> треугольное

[источник](https://dzen.ru/a/ZF-sTcTjmjcqIp4m)

```python
def u2t(elem):
    b_a = b - a
    if elem * (b - a) + a < c:
        return a + np.sqrt(b_a * (c - a) * elem)
    return b - np.sqrt(b_a * (b - c) * (1 - elem))

unif_dist = np.random.uniform(0, 1, n)
triang_dist = np.array(list(map(u2t, unif_dist)))

plt.hist(triang_dist, bins=bins_n, density=True)
plt.plot([a, c, b], [0, 2 / (b - a), 0])
plt.xlabel('x')
plt.ylabel('Плотность')
plt.show()
```

### Треугольное

```python
td2 = [rnd.triangular(a, b) for _ in range(n)]
plt.hist(td2, bins=bins_n, density=True)
plt.plot([a, c, b], [0, 2 / (b - a), 0])
plt.xlabel('x')
plt.ylabel('Плотность')
plt.show()
```
